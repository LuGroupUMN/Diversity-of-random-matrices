```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.insert(0, '../..') #hack
from train import train
from test import test
device = torch.device('cuda' if torch.cuda.is_available() else 'mps' if torch.backends.mps.is_available() else 'cpu')
print('Running on', device)
```

```python
d = 25
N = 40000
n = 40000
M = 10000
mlist = np.array([512, 768, 1152, 1728, 2592, 3888, 5832, 8748, 13122, 19683])
atype = 'const'
p = 0.5
lr = 0.001
num_epochs = 10000
```

```python
experiments = [
    [dict(train_method='fd',  test_method='fem', Vtype_train='piecewise', Vtype_test='piecewise', alpha_train=None, beta_train=None, alpha_test=None, beta_test=None)],
    [dict(train_method='fem', test_method='fd',  Vtype_train='piecewise', Vtype_test='piecewise', alpha_train=None, beta_train=None, alpha_test=None, beta_test=None)],
    [dict(train_method='fd',  test_method='fem', Vtype_train='piecewise', Vtype_test='lognormal', alpha_train=None, beta_train=None, alpha_test=2,    beta_test=2)],
    [dict(train_method='fd',  test_method='fem', Vtype_train='lognormal', Vtype_test='piecewise', alpha_train=2,    beta_train=2,    alpha_test=None, beta_test=None)],
    [dict(train_method='fd',  test_method='fem', Vtype_train='lognormal', Vtype_test='lognormal', alpha_train=2,    beta_train=2,    alpha_test=2,    beta_test=b) for b in [1, 3, 4]],
    [dict(train_method='fd',  test_method='fem', Vtype_train='lognormal', Vtype_test='lognormal', alpha_train=2,    beta_train=2,    alpha_test=a,    beta_test=2) for a in [1, 3, 4]],
]

all_results = []
for experiment in experiments:
    experiment_results = []
    for cfg in experiment:
        mdl = train(N, n, d, atype, cfg['Vtype_train'], p, cfg['alpha_train'], cfg['beta_train'], cfg['train_method'], lr, num_epochs, print_epoch=False)
        results = test(mdl, M, mlist, d, atype, cfg['Vtype_test'], p, cfg['alpha_test'], cfg['beta_test'], cfg['test_method'], error_type='relative')
        experiment_results.append((results, cfg))
    all_results.append(experiment_results)
```

```python
print(all_results)

y0s = [4e-2, 4.75e-2, 3e-2, 2e-2, 1.5e-2, 1.5e-2]
fig, axs = plt.subplots(2, 3, figsize=(12, 6), squeeze=False)
for ax, experiment_results, y0 in zip(axs.T.flat, all_results, y0s):
    for results, cfg in experiment_results:
        print([x - results[-1] for x in results])
        label = f"(α, β) = ({cfg['alpha_test']}, {cfg['beta_test']})"
        ax.plot(mlist[:-2], [x - results[-1] for x in results][:-2], 'o--', label=label)
    ax.plot(mlist[:-2], y0 * (mlist[:-2] / mlist[0])**-1, linestyle = '-', color='black', linewidth=2, alpha=0.5)
    ax.set_xlabel('m')
    if len(experiment_results) > 1:
        ax.legend()
    ax.loglog()
axs[0, 0].set_title('Discretization shift')
axs[0, 1].set_title('Shift in V(x)')
axs[0, 2].set_title('Shift in V(x)')
plt.tight_layout()
plt.show()
```

```python
def unsqueeze(value):
    if isinstance(value, list):
        return value
    return [value]

train_methods = ['fd', 'fd', 'fd', 'fd', ['fd']*3, ['fd']*3]
test_methods = ['fem', 'fd', 'fem', 'fem', ['fem']*3, ['fem']*3]
Vtype_trains = ['piecewise', 'piecewise', 'piecewise', 'lognormal', ['lognormal']*3, ['lognormal']*3]
Vtype_tests = ['piecewise', 'piecewise', 'lognormal', 'piecewise', ['lognormal']*3, ['lognormal']*3]
alpha_trains = [None, None, None, 2, [2]*3, [2]*3]
beta_trains  = [None, None, None, 2, [2]*3, [2]*3]
alpha_tests  = [None, None, 2, None, [2]*3, [1,3,4]]
beta_tests   = [None, None, 2, None, [1,3,4], [2]*3]
for i in range(6):
    train_methods[i] = unsqueeze(train_methods[i])
    test_methods[i] = unsqueeze(test_methods[i])
    Vtype_trains[i] = unsqueeze(Vtype_trains[i])
    Vtype_tests[i] = unsqueeze(Vtype_tests[i])
    alpha_trains[i] = unsqueeze(alpha_trains[i])
    beta_trains[i] = unsqueeze(beta_trains[i])
    alpha_tests[i] = unsqueeze(alpha_tests[i])
    beta_tests[i] = unsqueeze(beta_tests[i])
trim(1, train_methods, test_methods, Vtype_trains, Vtype_tests)
resultsss = [] # list of set of results
for ltrain_method, ltest_method, lVtype_train, lVtype_test, lalpha_train, lbeta_train, lalpha_test, lbeta_test in zip(train_methods, test_methods, Vtype_trains, Vtype_tests, alpha_trains, beta_trains, alpha_tests, beta_tests):
    resultsss.append([])
    for train_method, test_method, Vtype_train, Vtype_test, alpha_train, beta_train, alpha_test, beta_test in zip(ltrain_method, ltest_method, lVtype_train, lVtype_test, lalpha_train, lbeta_train, lalpha_test, lbeta_test):
        mdl = train(N, n, d, atype, Vtype_train, alpha_train, beta_train, train_method, lr, num_epochs)
        results = test(mdl, M, mlist, d, atype, Vtype_test, alpha_test, beta_test, test_method, error_type='relative')
        resultsss[-1].append(results)
```

```python
y0s = [1e-2, 4e-1, 1e-2, 1e-2, 1e-2, 1e-2]

fig, axs = plt.subplots(2, 3, figsize=(12, 6))
for i, (ax, graph, Vtype, lalpha, lbeta) in enumerate(zip(axs.T.flat, resultsss, Vtype_tests, alpha_tests, beta_tests)):
    for curve, alpha, beta in zip(graph, lalpha, lbeta):
        print([x - curve[-1] for x in curve])
        ax.plot(mlist, [x - curve[-1] for x in curve], 'o--', label = f'($\\alpha, \\beta$) = ({alpha}, {beta})')
        if len(lalpha)>1: ax.legend()
        ax.plot(mlist, y0s[i] * (mlist / mlist[0])**-1, label = 'O(-1)', linestyle = '-', color='black', linewidth=2)
        ax.loglog()

plt.tight_layout()
plt.show()
```
