```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.insert(0, '../..') #hack
from train import train
from test import test
device = torch.device('cuda' if torch.cuda.is_available() else 'mps' if torch.backends.mps.is_available() else 'cpu')
print('Running on', device)
```

```python
d = 25
N = 10000
n = 10000
M = 1000
mlist = np.array([512, 768, 1152, 1728, 2592, 3888, 5832, 8748, 13122, 19683])
atype = 'const'
method = 'fd'
p = 0.5
lr = 0.001
num_epochs = 4000
```

```python
Vtypes = ['piecewise', 'lognormal', 'lognormal', 'lognormal']
alphas = [0, 0, 1, 2]
betas = [1, 1, 2, 3]
mdls, resultss = [], []
for Vtype, alpha, beta in zip(Vtypes, alphas, betas):
    mdl = train(N, n, d, atype, Vtype, p, alpha, beta, method, lr, num_epochs)
    results = test(mdl, M, mlist, d, atype, Vtype, p, alpha, beta, method, error_type='mse')
    mdls.append(mdl)
    resultss.append(results)
```

```python
print(resultss, Vtypes, alphas, betas, sep='\n')
for results, Vtype, alpha, beta in zip(resultss, Vtypes, alphas, betas):
    plt.plot(mlist, results, 'o--', label = f'piecewise' if Vtype == 'piecewise' else f'($\\alpha, \\beta$) = ({alpha}, {beta})')

plt.plot(mlist, 3e-6 * (mlist / mlist[0])**-1, label = 'O(-1)', linestyle = '-', color='black', linewidth=2)
# plt.grid(True, alpha=0.3)
plt.loglog()
plt.legend()
plt.show()
```
